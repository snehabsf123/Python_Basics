Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?


---

Ans:


1. Overfitting
Definition:

Overfitting occurs when a model learns the training data too well, including noise and random fluctuations.
As a result, it performs very well on training data but poorly on new/unseen data.

Characteristics:

High training accuracy

Low testing accuracy

Model is too complex (too many parameters)

Example analogy:

Memorizing answers to specific questions instead of understanding the concepts

2. Underfitting
Definition:

Underfitting occurs when a model is too simple to capture the underlying patterns in the data.
It performs poorly on both training and testing data.

Characteristics:

Low training accuracy

Low testing accuracy

Model has high bias and low capacity

Example analogy:

Trying to fit a straight line to complex curved data.

4. How to Detect Overfitting & Underfitting
Detect Overfitting

Training accuracy very high, test accuracy low

Big gap between training and validation loss

Model performs well on training data but generalizes poorly

Detect Underfitting

Training accuracy is low

Validation/test accuracy is also low

Model fails to capture basic patterns

5. How to Prevent Overfitting
(a) Cross-Validation

Use k-fold cross-validation to ensure the model generalizes well.

Helps tune hyperparameters reliably.

(b) Regularization Techniques

L1 (Lasso) Regularization
Adds penalty = sum of absolute weights
‚Üí reduces weights to zero ‚Üí feature selection

L2 (Ridge) Regularization
Adds penalty = sum of squared weights
‚Üí reduces weight magnitude, prevents extreme values

Dropout (for Deep Learning)
Randomly drops neurons during training to prevent co-adaptation.

(c) Simplify the Model

Reduce number of parameters

Use pruning in tree-based models

(d) Early Stopping

Stop training when validation loss stops improving.

(e) Data Augmentation (in DL)

Add transformed copies of training data (images rotated, flipped, etc.).

(f) Get More Training Data

Larger datasets naturally reduce overfitting.

6. How to Prevent Underfitting

Increase model complexity (use more parameters/features)

Reduce regularization strength

Train longer (avoid early stopping too early)

Add more informative features

Use non-linear models instead of linear ones if needed

Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.


---

Ans:
 1. Deletion Methods (Removing Missing Data)

a) Listwise Deletion (Remove Entire Rows)

If a row has even one missing value, remove the entire row.

When to Use:

Dataset is large

Missing values are very few (<5%)

Missingness is random (MCAR: Missing Completely at Random)

Example:

Age	       Salary	    City

25	        50k	      Mumbai

NaN	        42k	      Delhi

30	        NaN	      Pune

Rows 2 and 3 would be deleted.

Pros: Simple, keeps dataset clean

Cons: Can cause major data loss

b) Column Deletion (Remove Entire Columns)

If a column has too many missing values, drop the column.

When to Use:

Column is not important

Column has >50‚Äì60% missing values

Example:

A ‚ÄúMiddle_Name‚Äù column with 80% missing values ‚Üí drop it.

2. Imputation Methods (Filling Missing Data)

a) Mean / Median / Mode Imputation

Replace missing values with basic statistical measures.

When to Use:

Mean ‚Üí numerical data (normally distributed)

Median ‚Üí numerical data (skewed distribution)

Mode ‚Üí categorical data

Example:

Dataset:

Age
22
24
NaN
23

Mean = (22+24+23)/3 = 23

Replace NaN ‚Üí 23.

b) KNN Imputation (K-Nearest Neighbor)

Fills missing values based on the most similar rows.

When to Use:

Numeric datasets

When relationships exist between features

For higher accuracy than simple mean/median

How it works:

Finds k closest rows

Uses their values to fill the missing data

Example: k=3 ‚Üí uses 3 most similar samples

c) Regression Imputation

Use a regression model to predict missing values.

Example:

To fill missing Salary, build a regression model:

Salary = f(Age, Experience, Education)


Predict missing Salary values using the trained model.

3. Advanced Methods

a) Multiple Imputation

Creates several imputed datasets

Trains model on each

Averages results for more robust predictions

Used in statistical modeling (e.g., R, SPSS).

b) Using Algorithms That Handle Missing Data Automatically

Some ML models can naturally deal with missing values:

XGBoost, LightGBM ‚Üí handle missing values internally

Decision Trees ‚Üí can split based on missing values

Example: XGBoost treats missing values as a separate branch.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).


---


Ans:  
An imbalanced dataset is one where the number of samples in each class is not evenly distributed.
One class (majority) has many more samples than the other (minority).

Example:

Fraud detection:

Legit transactions: 98%

Fraud transactions: 2%

The model may simply predict ‚Äúnot fraud‚Äù for everything and still get 98% accuracy‚Äîbut this is misleading.

Two Techniques to Handle Imbalanced Data
1. Resampling Techniques
(A) Oversampling

Increase minority class samples.

Theoretical idea:

Duplicate or synthetically generate new minority class samples

Helps balance class distribution

Practical methods:
a) Random Oversampling

Duplicate minority class examples.

b) SMOTE (Synthetic Minority Oversampling Technique)

Creates synthetic samples based on nearest neighbors, not duplicates.

Eg:
```
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_res, y_res = sm.fit_resample(X, y)

```

(B) Undersampling

Reduce majority class samples.

Theoretical idea:

Randomly remove samples from majority class

Useful when dataset is large

Eg:

```
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

```

2. Algorithm-Level Techniques

Instead of modifying the data, modify the model or loss function.

(A) Class Weighting

Assign higher weights to the minority class during training.

Theoretical idea:

The model penalizes misclassifying minority class more heavily.

Practical Example:

Eg:
```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X, y)

```
(B) Anomaly Detection / One-Class Classification

Useful when the minority class is extremely rare (fraud, faults).

Theoretical idea:

Model learns to detect outliers instead of classifying both classes equally.

Eg:

```
from sklearn.svm import OneClassSVM

model = OneClassSVM()
model.fit(normal_data)

```




Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.


---

Ans: Feature Scaling Is Important in ML

Feature scaling is the process of transforming numerical features to a similar scale (range or distribution).

Why it matters:

Many ML algorithms are distance-based or gradient-based, so features with large values can dominate those with small values.

Scaling helps to:

Improve model performance

Algorithms like KNN, SVM, and Logistic Regression rely on distances or gradients.

Speed up training

Gradient descent converges faster when features are scaled.

Prevent bias toward high-valued features

Features like ‚ÄúIncome = 50,000‚Äù may overshadow ‚ÄúAge = 30‚Äù.

Help algorithms treat all features equally

Algorithms that REQUIRE scaling:

KNN

K-Means

SVM

Logistic Regression

Neural Networks

PCA (Principal Component Analysis)

Algorithms that DON‚ÄôT need scaling much:

Tree-based models (Decision Trees, Random Forest, XGBoost)
(Because they split on threshold, not distance)

Min‚ÄìMax Scaling (Normalization)
Definition:

Rescales values to a fixed range, usually 0 to 1.


Characteristics:

Keeps the shape of the distribution

Sensitive to outliers (outliers stretch the range)

Good for algorithms requiring bounded values (e.g., Neural Networks with sigmoid/tanh)

Standardization (Z-score Normalization)

Characteristics:

Not bounded

Handles outliers better than Min-Max

Produces a standard normal distribution

Preferred for algorithms assuming normality (Logistic Regression, SVM, PCA)

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?


---

Ans:
Label Encoding

What it does:

Converts categorical values into numeric labels (integers).

Example:
["Red", "Blue", "Green"]
‚Üí
{"Red": 0, "Blue": 1, "Green": 2}

Characteristics:

Each category becomes a unique integer

Does not increase number of columns

Fast and memory-efficient

Problem:

It introduces a false ordinal relationship (e.g., ‚ÄúRed‚Äù < ‚ÄúBlue‚Äù), which can mislead algorithms that rely on numeric distances (SVM, KNN, Linear Models).

One-Hot Encoding
What it does:

Creates binary columns for each category.

Example:
["Red", "Blue", "Green"]
‚Üí

Red	Blue	Green

1	0	0

0	1	0

0	0	1

Characteristics:

No ordinal relationship

Good for models sensitive to numeric values

Increases dimensionality (may cause ‚Äúdummy variable explosion‚Äù)

When to Use Which?
‚úî Use Label Encoding When:

The categorical feature is ordinal
Examples:

Size ‚Üí Small < Medium < Large

Education ‚Üí High School < Bachelor < Master < PhD

You‚Äôre using tree-based models:

Decision Trees

Random Forest

XGBoost / LightGBM
These models do not get confused by integer labels.

The number of categories is very high (e.g., 10,000 unique values).
One-hot encoding would create too many columns.

‚úî Use One-Hot Encoding When:

The categorical feature is nominal (no order):
Colors, City, Country, Gender, etc.

You‚Äôre using distance-based or linear models that assume numeric meaning:

Logistic Regression

Linear Regression

KNN

SVM

Neural Networks

You want to avoid implying ordinal relationships.

Categories like EVENTS, EDUCATION, ART_AND_DESIGN, BOOKS_AND_REFERENCE, and PERSONALIZATION often appear among the highest-average-rating categories.

On the other hand, categories like GAME, AUTO_AND_VEHICLES, BEAUTY, and sometimes ENTERTAINMENT tend to have lower average ratings in some datasets

If you are analyzing such a dataset, category is a useful predictor / feature when trying to model rating or user satisfaction ‚Äî but must be complemented by other features (reviews count, installs, update recency, etc.) for robust insight.

````python
# Code to analyze category vs average rating

import pandas as pd

# Load dataset ‚Äî adjust path/filename as needed
df = pd.read_csv("googleplaystore.csv")

# Basic cleaning: drop rows without rating or without category
df = df.dropna(subset=['Rating', 'Category'])

# Convert rating to numeric (some may be string)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df = df.dropna(subset=['Rating'])

# Group by category, compute average rating and count of apps per category
cat_group = df.groupby('Category').agg(
    avg_rating = ('Rating', 'mean'),
    count_apps = ('App', 'count')
).reset_index()

# Sort by average rating descending
cat_sorted = cat_group.sort_values(by='avg_rating', ascending=False)

print(cat_sorted.head(10))
print(cat_sorted.tail(10))
````

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# LOAD DATA
# -------------------------
df = pd.read_csv("flight_price.csv")   # Adjust filename per your dataset folder

# Basic cleaning
df = df.dropna(subset=["price", "days_left", "airline", "source_city", "destination_city"])

# ====================================================
# PART A: PRICE vs DAYS LEFT (Dynamic Pricing Pattern)
# ====================================================

price_by_day = df.groupby("days_left")["price"].mean().reset_index()

print("Average Price by Days Left (first 10 rows):")
print(price_by_day.head(10))

# OPTIONAL PLOT
plt.figure(figsize=(10,5))
sns.lineplot(data=price_by_day, x="days_left", y="price")
plt.title("Average Flight Price vs Days Before Departure")
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()

# ====================================================
# PART B: Airline Comparison for Delhi ‚Üí Mumbai Route
# ====================================================

route_df = df[(df["source_city"].str.lower()=="delhi") & (df["destination_city"].str.lower()=="mumbai")]

airline_price = route_df.groupby("airline")["price"].mean().sort_values().reset_index()

print("\nAverage Prices for Delhi ‚Üí Mumbai Route:")
print(airline_price)

# OPTIONAL PLOT
plt.figure(figsize=(10,5))
sns.barplot(data=airline_price, x="airline", y="price")
plt.title("Airline Price Comparison for Delhi ‚Üí Mumbai")
plt.xticks(rotation=45)
plt.ylabel("Average Price")
plt.show()
````

Explanation & Insights

Part B: Number of projects

Employees with more projects tend to have slightly higher attrition.

Table of average projects by attrition helps quantify the effect.

 Advantages of this approach

Automatically detects correct column names to avoid KeyError or ValueError.

Works with numeric and categorical features.

Produces clear visualizations for reports.



````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# -------------------------
# LOAD DATA
# -------------------------
df = pd.read_csv("hr_analytics.csv")  # Adjust path if needed

# Inspect dataset
print("Columns in dataset:\n", df.columns)
print(df.head())

# -------------------------
# STEP 1: Map key columns safely
# -------------------------
def get_column(possible_keywords, df_columns):
    """Return the first matching column name, or None if not found."""
    for kw in possible_keywords:
        matches = [col for col in df_columns if kw.lower() in col.lower()]
        if matches:
            return matches[0]
    return None

attrition_col = get_column(['attrition','left'], df.columns)
satisfaction_col = get_column(['satisfaction'], df.columns)
salary_col = get_column(['salary','income'], df.columns)
overtime_col = get_column(['overtime','ot'], df.columns)
num_projects_col = get_column(['project'], df.columns)
worklife_col = get_column(['worklife','wlb'], df.columns)
years_col = get_column(['year','company','tenure'], df.columns)

print("\nMapped columns:")
print(f"Attrition: {attrition_col}, Satisfaction: {satisfaction_col}, Salary: {salary_col}, Overtime: {overtime_col}, Projects: {num_projects_col}, WorkLife: {worklife_col}, Years: {years_col}")

# -------------------------
# PART A: Factors correlated with attrition
# -------------------------
numeric_cols = df.select_dtypes(include=['int64','float64']).columns
corr_matrix = df[numeric_cols].corr()

if attrition_col in corr_matrix.columns:
    attrition_corr = corr_matrix[attrition_col].sort_values(ascending=False)
    print("\nCorrelation of numeric features with Attrition:")
    print(attrition_corr)

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Key features visualization
key_features = [satisfaction_col, overtime_col, salary_col, worklife_col, years_col]
for feature in key_features:
    if feature is None:
        continue  # Skip missing columns
    plt.figure(figsize=(6,4))
    if df[feature].dtype == 'object':
        sns.countplot(data=df, x=feature, hue=attrition_col)
    else:
        sns.boxplot(data=df, x=attrition_col, y=feature)
    plt.title(f"{feature} vs {attrition_col}")
    plt.show()

# -------------------------
# PART B: Number of projects vs attrition
# -------------------------
if num_projects_col:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x=attrition_col, y=num_projects_col)
    plt.title("Number of Projects vs Attrition")
    plt.show()

    proj_avg = df.groupby(attrition_col)[num_projects_col].mean().reset_index()
    print("\nAverage Number of Projects by Attrition:")
    print(proj_avg)
````
