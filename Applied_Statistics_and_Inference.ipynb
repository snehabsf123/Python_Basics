Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?



---

Ans: In hypothesis testing, we make decisions about a population using sample data. Two kinds of errors can occur:
1) Type I Error (False Positive)

Rejecting the null hypothesis (H₀) when it is actually true.


Concluding that an effect or difference exists when it really does not.

Eg:
Critical in fields like medicine or law, where claiming an effect that doesn’t exist can be costly.

2) Type II Error (False Negative)

Failing to reject the null hypothesis (H₀) when the alternative hypothesis (H₁) is true.

Missing a real effect or difference.

Eg:
Important in quality control, safety testing, and medical screening where failing to detect a real issue is harmful.





Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?


---

Ans: The P-value is the probability of obtaining a sample result as extreme as or more extreme than the one observed, assuming the null hypothesis (H₀) is true.

To Interpret the P-value

1. Small P-value (typically ≤ 0.05)

Indicates that the observed data would be very unlikely if H₀ were true.

Provides evidence against the null hypothesis.

Decision: Reject H₀.

2. Large P-value (> 0.05)

Observed data are consistent with what we would expect under H₀.

Not strong enough evidence to reject H₀.

Decision: Fail to reject H₀



Question 3:Explain the difference between a Z-test and a T-test, including when to use
each.


---

Ans: Z-tests and T-tests are both used to test hypotheses about population means, but they differ based on sample size and knowledge of population variance.

1. Z-test

Large sample size (typically n ≥ 30).

Population standard deviation (σ) is known.

Data are approximately normally distributed (or sample is large enough for the Central Limit Theorem to apply).

Applications:

Testing population mean with known variance

Comparing two proportions

Quality control, industrial settings

2. T-test

Small sample size (n < 30).

Population standard deviation (σ) is unknown — must estimate it using sample standard deviation (s).

Data should be approximately normally distributed.

Applications:

Testing whether the average height of a small group differs from the national average.

Checking if the average test score of a class differs from the school average




Question 4:What is a confidence interval, and how does the margin of error influence
its width and interpretation?


---

Ans:

* A confidence interval (CI) is a range of values, calculated from sample data, that is likely to contain the true population parameter (like the mean or proportion) with a specified confidence level.

Examples of confidence levels:

90% confidence interval

95% confidence interval

99% confidence interval

A 95% confidence interval means that if we repeated the sampling process many times, 95% of the intervals constructed would contain the true population parameter.
*   The margin of error (MOE) is the amount added and subtracted from the sample estimate to create the confidence interval.

Margin of error depends on:

Sample size (n)

Confidence level (Z or t-value)

Variability in data (standard deviation)



*   Examples


Narrow CI (small margin of error)

95% CI: 50 ± 2 → (48, 52)

More precise estimate.

Wide CI (large margin of error)

95% CI: 50 ± 10 → (40, 60)

Less precise estimate.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?


---

Ans:


*   Purpose of ANOVA

The purpose of ANOVA is to determine whether there are significant differences in the means of three or more groups.

Instead of running multiple t-tests (which increases error), ANOVA allows us to test all groups at once with a single hypothesis test.



*   Assumptions of ANOVA

ANOVA is based on three key assumptions:

The data in each group must be collected independently.

The data within each group should be approximately normally distributed.

All groups should have roughly equal variances (similar spread of data).



*   ANOVA Extends Hypothesis Testing Beyond Two Groups

A t-test can only compare two group means.

ANOVA extends this idea by comparing three or more groups at the same time.

Example: 3 groups → 3 t-tests → higher chance of a wrong conclusion.

How ANOVA solves this:

ANOVA compares variability between groups and variability within groups using the F-statistic.

It tests whether the differences between group means are larger than expected by chance.





Question 6: Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset.


---



```python
import numpy as np
from scipy.stats import norm

# ----------------------------
# Sample Data
# ----------------------------
data = [52, 55, 48, 50, 49, 53, 54, 47, 51, 50]  # Sample dataset
population_mean = 50      # H0: μ = 50
population_std = 5        # Known population standard deviation
alpha = 0.05              # Significance level

# ----------------------------
# Z-test Calculations
# ----------------------------
sample_mean = np.mean(data)
n = len(data)

# Z statistic
z_value = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_value)))

# ----------------------------
# Results
# ----------------------------
print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference detected.")
```

Question 7:Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.


---



```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Simulate Binomial Data
# ----------------------------
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of observations

data = np.random.binomial(n, p, size)

# ----------------------------
# Plot Histogram
# ----------------------------
plt.hist(data, bins=range(0, n+2), edgecolor='black', alpha=0.7)
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.xticks(range(0, n+1))
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```

Question 8: Generate multiple samples from a non-normal distribution and implement
the Central Limit Theorem using Python.


---



```python
import numpy as np
import matplotlib.pyplot as plt

population = np.random.exponential(scale=1, size=100000)

sample_size = 30          # n for each sample
num_samples = 1000        # number of samples

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(15,5))

# Plot 1: Histogram of Original Non-Normal Population
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black')
plt.title("Original Population (Exponential Distribution)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plot 2: Histogram of Sample Means (CLT)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black')
plt.title("Distribution of Sample Means (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Question 9: Write a Python function to calculate and visualize the confidence interval
for a sample mean.


---



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):

    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error

    # t-score for the given confidence level
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of Error
    margin_error = t_score * std_err

    # Confidence Interval
    lower_bound = mean - margin_error
    upper_bound = mean + margin_error

    print(f"Sample Mean: {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.3f}, {upper_bound:.3f})")


    plt.figure(figsize=(6, 4))
    plt.errorbar(x=0, y=mean, yerr=margin_error, fmt='o', color='blue', capsize=10)
    plt.xlim(-1, 1)
    plt.title(f"{int(confidence*100)}% Confidence Interval for the Sample Mean")
    plt.xticks([])
    plt.ylabel("Value")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

    return mean, lower_bound, upper_bound


sample_data = [52, 55, 48, 50, 49, 53, 54, 47, 51, 50]
confidence_interval(sample_data, confidence=0.95)
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare
observed and expected distributions, and explain the outcome.


---



```python
import numpy as np
from scipy.stats import chisquare

observed = np.array([8, 9, 10, 11, 12, 10])  # observed counts for each face


expected = np.array([10, 10, 10, 10, 10, 10])  # 60 rolls / 6 faces


chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)


print("Chi-square Statistic:", chi2_stat)
print("P-value:", p_value)


alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution significantly differs from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected distributions.")
```
