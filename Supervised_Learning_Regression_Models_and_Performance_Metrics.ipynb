Logistic Regression predicts a probability or class (e.g., whether a house will sell or not).

Output Range:

Linear Regression can output any real number.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.


---

Ans:

```python
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Predictor (must be 2D)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])     # Response

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Get slope (coefficient) and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Î²1): {slope}")
print(f"Intercept (Î²0): {intercept}")
```
